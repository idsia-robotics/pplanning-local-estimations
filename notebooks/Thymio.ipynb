# Mighty Thymio Playground

This notebook illustrates how to compute a plan for the Mighty Thymio

```python
from ple.map import ThymioMap
from ple.thymio_planner import Plan
```

```python
thymio_map = ThymioMap(image_path="../maps/thymio_test.npy", 
                       model_path='../models/thymio.h5', 
                       size=5.0)
```

```python
plt.figure(figsize=(10,10))
thymio_map.plot()
```

We constrains to move forwards at 8 cm/s and solve the planning problem for 60 seconds.

```python
plan = Plan(thymio_map, pose=(1, 4, 0), target=(4, 1.5, 0), 
            tolerance=0.2, threshold=0.9, allow_moving_backward=False)
plan.solve(60)
```

```python
plt.figure(figsize=(10,10))
plan.plot()
thymio_map.plot()
```

```python
from matplotlib import pyplot as plt
xyt = plan.curve(0.5, ros=False)
plt.figure(figsize=(10,10))
plt.plot(xyt[:,0], xyt[:,1], '-')
plt.axis('equal')
thymio_map.plot()
```

We let the robot move backwards too:

```python
plan_1 = Plan(thymio_map, pose=(1, 4, 0), target=(4, 1.5, 0), tolerance=0.2, threshold=0.9,allow_moving_backward=True)
plan_1.solve(60)
```

```python
plt.figure(figsize=(10,10))
plan_1.plot()
thymio_map.plot()
```
